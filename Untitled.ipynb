# Financial Performance Analysis of Global Energy Sector (2018-2025)

## 1. Problem Definition
The energy industry is a capital-intensive sector characterized by high volatility and significant market cycles. This project aims to build an interactive data product to evaluate the financial health and operational efficiency of global energy giants (e.g., ExxonMobil, BP). 

We utilize the **DuPont Analysis framework** to decompose **Return on Assets (ROA)** into two levers:
1. **Profitability** (Net Profit Margin)
2. **Efficiency** (Asset Turnover)

By visualizing these metrics, the tool helps stakeholders identify whether a company's performance is driven by high-profit margins or superior asset productivity.

```python
import wrds
import pandas as pd

def download_energy_data_from_wrds(username):
    """
    自动登录 WRDS 并抓取能源行业上市公司财务数据
    """
    try:
        # 1. 建立连接 (首次运行会提示你输入密码并询问是否保存登录令牌)
        db = wrds.Connection(wrds_username=username)
        
        print("--- 正在执行 SQL 查询 (Compustat North America) ---")
        
        # 2. 编写 SQL 查询语句
        # sic 1311: 原油和天然气开采 | sic 2911: 石油精炼 | sic 4911: 电力服务
        # --- 修正后的 SQL 查询语句 ---
        # --- 修正后的 SQL 查询语句 ---
        sql_query = """
        SELECT 
            tic, conm, fyear, sale, cogs, ni, at, sich
        FROM 
            comp.funda
        WHERE 
            indfmt = 'INDL'           -- 工业格式
            AND datafmt = 'STD'       -- 标准化数据
            AND popsrc = 'D'          -- 国内源码
            AND consol = 'C'          -- 合并报表
            AND fyear BETWEEN 2018 AND 2025  -- 时间范围
            -- 修正字段名为 sich
            AND sich IN ('1311', '2911', '4911') 
        """
        
        # 3. 执行查询并转换为 Pandas DataFrame
        df = db.raw_sql(sql_query)
        
        # 4. 数据预处理
        # 确保数值字段没有干扰字符
        cols = ['sale', 'cogs', 'ni', 'at']
        df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
        
        # 5. 保存数据
        output_file = "wrds_energy_data.csv"
        df.to_csv(output_file, index=False)
        
        print(f"--- 下载成功！文件已保存为: {output_file} ---")
        print(f"数据量: {len(df)} 行")
        
        # 关闭连接
        db.close()
        return df

    except Exception as e:
        print(f"发生错误: {e}")
        return None

# --- 执行区 ---
# 替换为你的真实 WRDS 用户名
MY_WRDS_USERNAME = 'your_username_here' 

if __name__ == "__main__":
    energy_df = download_energy_data_from_wrds(MY_WRDS_USERNAME)
    if energy_df is not None:
        print(energy_df.head())
```

## 3. Python Implementation: Data Processing & Business Insights
